## <span style="color:blue">Immutable and Mutable Objects</span>
+ 字串，數值，tuple是immutable object

```python
list1 = [1,2,3,4]
list2 = list1
list2[1] = 5
print(list1)
print(list2)
```

## <span style="color:blue">函式</span>
語法:<br>
def functionName(par1,par2,par3=defaultValue,par4=defaultValue2):<br>
>indented block of statememts<br>
>return expression

呼叫函式時par1,par2照順序傳遞(positional parameters)，par3與par4為keyword parameter，可以不照順序，但是要寫參數名稱。

```python
def mystery(a,b,c=1):
    return a+b*c

print(mystery(10,2))
print(mystery(10,2,c=3))
```

若是函式欲使用全域變數，在函式裡用global宣告變數為全域變數。

```python
x=5
y=1
def variableScope():
    print(x)
    # x=1 #若這行沒變成註解，x會被視為local variable，這個程式會產生UnboundLocalError
    z=y #可以讀global variable y
    print(z)
    return  

def globalVariable():
    global x,y #直接宣告x,y為全域變數，清楚明確
    print(x)
    x=1
    z=y
    print(z)
    return

variableScope()
globalVariable()
```

## <span style="color:blue"> 檔案輸出 </span>

### 步驟
+ 開啟輸出檔: open
+ 輸出: print/write 
+ 關閉檔案: close 

### 如C一般的寫法

```python
out = open('myfile.txt','w')
for i in range(100):
    print(i,file=out)
out.close()
```

### 使用with敘述，在離開with的區塊會自動關閉檔案

```python
with open('myfile.txt','w') as out:
    for i in range(100):
        print(i,file=out)
```

### 可以使用write instance method取代print

```python
with open('myfile.txt','w') as out:
    for i in range(100):
        out.write(str(i)+'\n')
```

## <span style="color:blue"> 檔案輸入 </span>

```python
infp = open('myfile.txt','r')
A = []
while True:
    a= infp.readline().strip()
    if a:
        A.append(a)
    else:
        break
infp.close()
print(len(A),A)
```

```python
A=[]
with open('myfile.txt','r') as infp:
    for a in infp.readlines():
        A.append(a.strip())
print(len(A),A)
```
